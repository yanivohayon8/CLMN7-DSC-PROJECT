```python
import pandas as pd
import glob
```

```python
experiments_path = '../data/processed/'
csv_files = glob.glob(experiments_path + '/**/*.csv',recursive=True)
```

```python
dfs = [pd.read_csv(csv) for csv in csv_files]
```

```python
all_dfs = pd.concat(dfs)
```
