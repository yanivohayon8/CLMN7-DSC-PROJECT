```python
from abc import ABC
#import gensim.corpora as corpora
#import gensim
import numpy as np



class CalculateVectors(ABC):
    '''
        2.	Calculate vector 
            a.	Tf-idf , TF, Word embedding 
            b.	Make the calculation across all text (like word embedding) or
                just within the block or adjacent block (what ever you would like)
                Choose sliding window  fixed size or documents
    '''
    
    blocks = NotImplemented
    
    def calc(self):
        pass


    
'''
    consider aggregate the 2 classes into single class
'''    
    
from sklearn.feature_extraction.text import CountVectorizer
class TFCalculator(CalculateVectors):
    def __init__(self,blocks): 
        self.blocks = blocks
        self.vectorizer = CountVectorizer()
        
    
    # each vector is n word in all of the vocabulary
    def calc(self):
        #print(self.blocks)
        con_blocks = [' '.join(blk) for blk in self.blocks]
        #print(con_blocks)
        return self.vectorizer.fit_transform(con_blocks).toarray()
        
from sklearn.feature_extraction.text import TfidfVectorizer
class TFIDFCalculator(CalculateVectors):
    def __init__(self,blocks):
        self.blocks = blocks
        self.vectorizer = TfidfVectorizer()
    
    def calc(self):
        con_blocks = [' '.join(blk) for blk in self.blocks]
        return self.vectorizer.fit_transform(con_blocks).toarray()
```

# Word embedding(wikipedia), cosine + tf-idf,

```python
we_block_handler = CreateBlocks(transcripts_jsons,window_size=30)
we_blocks = word_embedding_block_handler.partion()
```

```python
we_tf_idf_calculator = TFIDFCalculator(word_embedding_blocks)
we_vectors_matrix = we_tf_idf_calculator.calc()
```

```python
tf_idf_df = pd.DataFrame(we_vectors_matrix,columns= we_tf_idf_calculator.vectorizer.get_feature_names())
tf_idf_df.describe()
```

```python
we_mat = [] #np.ndarray(len(word_embedding_blocks))
for i in range(len(word_embedding_blocks)):
    block_vector = []
    for j in range(len(word_embedding_blocks[i])):
        try:
            word_vect = wiki[word_embedding_blocks[i][j]]
            word_vect = word_vect * tf_idf_df[word_embedding_blocks[i][j]][i]
            block_vector.append(word_vect)
        except Exception:
            pass
    we_mat.append(sum(block_vector)/len(block_vector))
```

```python
print(len(we_mat))
#print(we_mat[0])
we_mat_similarity = cosine_similarity(we_mat,we_mat)
```

```python
MyPlotting.similarity_matrix(we_mat_similarity)
```
