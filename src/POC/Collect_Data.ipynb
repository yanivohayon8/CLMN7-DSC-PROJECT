```python
from youtube_transcript_api import YouTubeTranscriptApi
import urllib.parse as urlparse
from pytube import YouTube
import cv2
import os
import glob
```

```python
def get_transcript(url):
    url_data = urlparse.urlparse(url)
    query = urlparse.parse_qs(url_data.query)
    video_id = query["v"][0]
    print ("The video %s ID is %s" % (url,video_id))
    return video_id
```

```python
# extract transcript
video_id = get_transcript(curr_url)
t = YouTubeTranscriptApi.get_transcript(video_id)
```

```python
def extract_images_from_video(video, delay=30, name="file", max_images=20, silent=False):    
    vidcap = cv2.VideoCapture(video)
    count = 0
    num_images = 0
    folder = os.getcwd()
    label = 0
    success = True
    fps = int(vidcap.get(cv2.CAP_PROP_FPS))
    
    while success and num_images < max_images:
        success, image = vidcap.read()
        num_images += 1
        label += 1
        file_name = name + "_" + str(label) + ".jpg"
        path = os.path.join(folder, file_name)
        cv2.imwrite(path, image)
        if cv2.imread(path) is None:
            os.remove(path)
        else:
            if not silent:
                print(f'Image successfully written at {path}')
        count += delay*fps
        vidcap.set(1, count)
```

```python
for i, video in enumerate(glob.glob("*.mp4")):
        extract_images_from_video(video,  delay=30, name="yalla", max_images=1500, silent=False)
```
